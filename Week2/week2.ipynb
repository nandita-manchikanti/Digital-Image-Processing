```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('Lena.png',0)
plt.imshow(img,cmap='gray')
plt.show()
```

```python
img1=cv2.resize(img,(128,128))
img2=cv2.resize(img,(64,64))
img3=cv2.resize(img,(32,32))
img4=cv2.resize(img,(16,16))

fig, ax = plt.subplots(1,4,figsize=(17,17))
ax[0].imshow(img1,cmap='gray'),ax[0].set_title('128x128')
ax[1].imshow(img2,cmap='gray'),ax[1].set_title('64x64')
ax[2].imshow(img3,cmap='gray'),ax[2].set_title('32x32')
ax[3].imshow(img4,cmap='gray'),ax[3].set_title('16x16')
```

When the number of pixels per unit area decreases we observe Aliasing.
